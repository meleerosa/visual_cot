```python
import json

path = "/root/project/data/visual_genome/relationships.json"

with open(path, "r") as f:
    data = json.load(f)

# 상위 몇 개만 출력
for i, item in enumerate(data[:5]):
    print(f"--- Entry {i} ---")
    print("Image ID:", item.get("image_id"))
    print("Relationships:", item.get("relationships", []))
    print()
```

```python
from collections import Counter

def extract_all_predicates(data):
    predicate_counter = Counter()

    for item in data:
        for rel in item.get('relationships', []):
            predicate = rel.get('predicate', '').strip().lower()
            if predicate:
                predicate_counter[predicate] += 1

    return predicate_counter
```

```python
predicate_counts = extract_all_predicates(data)

# 상위 20개 출력
print("🔍 Predicate 종류 및 빈도:")
for predicate, count in predicate_counts.most_common(1000):
    print(f"{predicate}: {count}")
```

```python
import spacy
from tqdm import tqdm
from spacy.symbols import VERB, ADP, NOUN

nlp = spacy.load("en_core_web_sm")

def classify_predicate(predicate):
    doc = nlp(predicate.lower().strip())
    if not doc:
        return "OTHER"

    if any(token.pos_ == "VERB" for token in doc):
        return "VERB"
    elif any(token.pos_ == "ADP" for token in doc):
        return "PREP_PHRASE"
    elif any(token.pos_ == "NOUN" for token in doc):
        return "NOUN"
    else:
        return "OTHER"

def conjugate_verb(subject, verb):
    irregulars = {'have': 'has', 'be': 'is', 'wear': 'wears', 'carry': 'carries', 'hold': 'holds'}
    verb = verb.lower()
    if verb in irregulars:
        return irregulars[verb]
    if subject.lower().startswith("the ") and subject.split()[1].endswith("s"):
        return verb
    if verb.endswith('y'):
        return verb[:-1] + 'ies'
    elif verb.endswith(('sh', 'ch', 'x', 's', 'z')):
        return verb + 'es'
    else:
        return verb + 's'

def generate_sentence(subject, predicate, obj):
    category = classify_predicate(predicate)
    s = f"The {subject.strip()}"
    o = f"the {obj.strip()}"
    p = predicate.strip().lower()

    if category == "VERB":
        verb = conjugate_verb(s, p)
        return f"{s} {verb} {o}"
    elif category == "PREP_PHRASE":
        return f"{s} is {p} {o}"
    elif category == "NOUN":
        return f"{s} has the {p}"
    else:
        return f"{s} is {p} {o}"

def process_visual_genome(data):
    image_sentences = {}

    for image in tqdm(data):
        image_id = image.get('image_id', 'unknown')
        relationships = image.get('relationships', [])

        sentences = set()  # Avoid duplicates
        for rel in relationships:
            subj = rel.get('subject', {}).get('name', '').strip()
            pred = rel.get('predicate', '').strip()
            obj = rel.get('object', {}).get('name', '').strip()

            if subj and pred and obj:
                sentence = generate_sentence(subj, pred, obj)
                sentences.add(sentence)

        if sentences:
            image_sentences[image_id] = list(sentences)

    return image_sentences
```
